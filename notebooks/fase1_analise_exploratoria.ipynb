# Configuração do ambiente

```python
# Importação de bibliotecas para manipulação e visualização de dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para melhorar a visualização dos gráficos
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (15, 8)
plt.rcParams['font.size'] = 12

print("Ambiente configurado e bibliotecas importadas com sucesso!")
```

# Amostra dos dados

```python
# Carregando os dados. Substitua 'seu_dataset.csv' pelo nome do seu arquivo.
try:
    df = pd.read_csv('../data/bank_transactions_data.csv')
    print("Dataset carregado com sucesso!")
    print(f"O dataset contém {df.shape[0]} linhas e {df.shape[1]} colunas.")
except FileNotFoundError:
    print("ERRO: Arquivo não encontrado. Verifique o caminho e nome do arquivo.")

# --- Inspeção Inicial ---

# 1. Visualizar as primeiras linhas
print("\n--- Amostra dos Dados ---")
display(df.head())

# 2. Verificar tipos de dados e valores nulos
print("\n--- Informações Gerais e Tipos de Dados ---")
df.info()

# 3. Obter resumo estatístico das variáveis numéricas
print("\n--- Resumo Estatístico ---")
display(df.describe())

```

# Análise de Variáveis Numéricas

```python
# Criando diretório para salvar gráficos
import os
plots_dir = '../results/plots'
os.makedirs(plots_dir, exist_ok=True)

# Histograma do Valor da Transação (TransactionAmount)
plt.figure(figsize=(12, 6))
sns.histplot(df['TransactionAmount'], bins=50, kde=True)
plt.title('Distribuição do Valor das Transações')
plt.xlabel('Valor da Transação')
plt.ylabel('Frequência')
plt.tight_layout()
plt.savefig(f'{plots_dir}/distribuicao_valor_transacoes.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"✅ Gráfico salvo em: {plots_dir}/distribuicao_valor_transacoes.png")

# Histograma da Idade do Cliente (CustomerAge)
plt.figure(figsize=(12, 6))
sns.histplot(df['CustomerAge'], bins=30, kde=True)
plt.title('Distribuição da Idade dos Clientes')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.tight_layout()
plt.savefig(f'{plots_dir}/distribuicao_idade_clientes.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"✅ Gráfico salvo em: {plots_dir}/distribuicao_idade_clientes.png")

```

## Qual é a faixa de valor mais comum para uma transação?
Analisando o histograma "Distribuição do Valor das Transações":

Concentração em Valores Baixos: A grande maioria das transações, indicada pelas barras mais altas do gráfico, está concentrada na extremidade esquerda. Isso mostra que a faixa de valor mais comum é para transações de baixo valor.

Faixa Mais Comum: Observando o eixo "Valor da Transação", a frequência mais alta ocorre aproximadamente na faixa de R$ 0 a R$ 250. A maior parte das operações financeiras neste dataset se enquadra nesse intervalo.

Cauda Longa à Direita (Long Tail): O gráfico mostra uma "cauda longa" que se estende para a direita. Isso significa que, embora muito menos frequentes, existem transações com valores bem mais altos (R$ 500, R$ 1000, e até acima de R$ 1500), mas elas são raras em comparação com as transações de baixo valor.

### Conclusão para a Questão 1: 
A faixa de valor mais comum para uma transação é, sem dúvida, a de valores baixos, principalmente entre R$ 0 e R$ 250. Essa é a característica dominante do comportamento "normal" das transações no que diz respeito ao seu valor.


## Qual é a faixa etária principal dos clientes?
A distribuição é bimodal, o que significa que existem duas faixas etárias principais que se destacam no dataset:

Clientes Jovens: O pico mais alto de frequência ocorre para clientes na faixa dos 18 aos 30 anos. Este é o grupo mais numeroso de clientes que realizam transações.

Clientes Mais Velhos: Há um segundo pico, também muito relevante, para clientes na faixa dos 50 aos 60 anos.

### Conclusão para a Questão 2: 
Não há uma única faixa etária principal, mas sim duas. O comportamento "normal" do cliente neste dataset é predominantemente de pessoas mais jovens (18-30) e de pessoas mais maduras (50-60), com uma frequência menor de clientes na faixa intermediária (35-45 anos).

# Análise de Variáveis Categóricas

```python
# Gráfico de Contagem para o Canal da Transação (Channel)
plt.figure(figsize=(10, 6))
sns.countplot(x='Channel', data=df, order=df['Channel'].value_counts().index)
plt.title('Contagem de Transações por Canal')
plt.xlabel('Canal')
plt.ylabel('Número de Transações')
plt.tight_layout()
plt.savefig(f'{plots_dir}/contagem_transacoes_canal.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"✅ Gráfico salvo em: {plots_dir}/contagem_transacoes_canal.png")

# Gráfico de Contagem para a Ocupação do Cliente (CustomerOccupation)
plt.figure(figsize=(14, 8))
sns.countplot(y='CustomerOccupation', data=df, order=df['CustomerOccupation'].value_counts().index)
plt.title('Contagem de Transações por Ocupação do Cliente')
plt.xlabel('Número de Transações')
plt.ylabel('Ocupação')
plt.tight_layout()
plt.savefig(f'{plots_dir}/contagem_transacoes_ocupacao.png', dpi=300, bbox_inches='tight')
plt.show()
print(f"✅ Gráfico salvo em: {plots_dir}/contagem_transacoes_ocupacao.png")

print(f"\n🎉 Todos os gráficos da análise exploratória salvos em: {plots_dir}/")

```

## Qual o canal mais utilizado ?

Distribuição Equilibrada: Diferente das outras análises, aqui não há um canal que domine completamente. Os três canais ("Branch" - Agência, "ATM" e "Online") têm um volume de transações muito similar.

Canal Mais Utilizado: Embora a diferença seja pequena, o canal "Branch" (Agência) é o mais utilizado, com a barra mais alta. Logo em seguida vêm "ATM" e "Online", ambos com uma contagem quase idêntica.

### Conclusão para a Questão 3:
O comportamento "normal" das transações é bem distribuído entre os canais físicos e digitais. O canal Branch (Agência) é ligeiramente o mais frequente, mas ATM e Online são praticamente tão comuns quanto. Isso indica que os clientes deste dataset utilizam uma variedade de métodos para realizar suas operações financeiras, sem uma preferência avassaladora por um único canal.
